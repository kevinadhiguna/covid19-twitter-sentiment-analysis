## Twitter API Credentials

```
import tweepy

consumerKey = 'YOUR CONSUMER KEY'
consumerSecret = 'YOUR CONSUMER SECRET'
accessToken = 'YOUR ACCESS TOKEN'
accessTokenSecret = 'YOUR ACCESS TOKEN SECRET'

authenticate = tweepy.OAuthHandler(consumerKey, consumerSecret)
authenticate.set_access_token(accessToken, accessTokenSecret)
api = tweepy.API(authenticate, wait_on_rate_limit=True)
```

```
posts = api.search(
    q = "#PSBB OR ((#covid19 OR covid19) AND (indonesia OR pemerintah)) OR ((#wfh OR wfh) AND (indonesia OR pemerintah)) OR PSBB", 
    count = 100
  )

i = 1

print("Show 10 recent tweets : \n")
for tweet in posts[0:10]:
  print(str(i) + ') ' + tweet.text + '\n')
  i = i+1
```

## Data Preprocessing and Sentiment Analysis

```
import re
from textblob import TextBlob

analysisResult = []

for tweet in posts:
  tweet_properties = {}
  tweet_properties['date'] = tweet.created_at
  tweet_properties['text'] = tweet.text
  tweet_properties['username'] = tweet.user.screen_name
  tweet_properties['location'] = tweet.user.location
  clean_tweets = ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)"," ",tweet.text).split())

  analysis = TextBlob(clean_tweets)

  if analysis.sentiment.polarity > 0.0:
    tweet_properties['sentiment'] = 'Positive'
  elif analysis.sentiment.polarity == 0.0:
    tweet_properties['sentiment'] = 'Neutral'
  else:
    tweet_properties['sentiment'] = 'Negative'

  if tweet.retweet_count > 0:
    if tweet_properties not in analysisResult:
      analysisResult.append(tweet_properties)
  else:
    analysisResult.append(tweet_properties)
```

```
analysisResult
```

### Number of each Tweets

```
positive_tweets = [t for t in analysisResult if t["sentiment"]=="Positive"]

negative_tweets = [t for t in analysisResult if t["sentiment"]=="Negative"]

neutral_tweets = [t for t in analysisResult if t["sentiment"]=="Neutral"]

print("Number of Positive Tweets : " + str(len(positive_tweets)))
print("Number of Negative Tweets : " + str(len(negative_tweets)))
print("Number of Neutral Tweets : " + str(len(neutral_tweets)))
```

## Sentiment Result

```
print("Positive Tweets: ", len(positive_tweets), "({} %)".format(100*len(positive_tweets)/len(analysisResult)))

print("Negative Tweets: ", len(negative_tweets), "({} %)".format(100*len(negative_tweets)/len(analysisResult)))

print("Neutral Tweets: ", len(neutral_tweets), "({} %)".format(100*len(neutral_tweets)/len(analysisResult)))
```
